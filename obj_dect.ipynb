# 1. Inference

`detect.py` runs inference on a variety of sources, downloading models automatically, and saving results to `runs/detect`. Example inference sources are:

```shell
python detect.py --source 0  # webcam
                          img.jpg  # image 
                          vid.mp4  # video
                          path/  # directory
                          path/*.jpg  # glob
                          'https://youtu.be/Zgi9g1ksQHc'  # YouTube
                          'rtsp://example.com/media.mp4'  # RTSP, RTMP, HTTP stream
```

````python
!python detect.py --weights yolov5s.pt --img 640 --conf 0.25 --source "https://image3.mouthshut.com/images/Restaurant/Photo/-47307_119006.jpg"
display.Image(filename='runs/detect/exp/-47307_119006.jpg', width=600)
````
